# Using the given MT cars dataset you need to find outlier with z score.

```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv('/content/mtcars.csv')
df
```

```
outliers=[]
def detect_outliers(data):
  t=2
  mean=np.mean(data)
  std=np.std(data)
  for i in data:
    z_score=(i-mean)/std
    if np.abs(z_score)>t:
      outliers.append(i)
  return outliers
```
